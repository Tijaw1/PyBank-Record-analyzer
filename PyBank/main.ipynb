```python
# HW2: python script to analyze record of data for pybank 
# Importing numpy 
import numpy as np
# Importing pathlib and csv file
from pathlib import Path
import csv
# File path def
csvpath = Path ('C:/Users/Endalkachew Desta/Downloads/HW2_PyBank_Resources_budget_data.csv')
# Initializing variables dates and profitLosses
dates = []
profitLosses = []
lineNum = 0
# Opening the input as csv object 
with open(csvpath,'r') as csvfile:
    csvdata=csv.reader(csvfile, delimiter = ',')
       
    header = next(csvdata)
    lineNum +=1
    
    print(f"{header} <---- Header")
    
    for row in csvdata:
        print(row)
        
        profit = int(row[1])
        date = row[0]
        dates.append(date)
        profitLosses.append(profit)
# Initializing                 
max_profit_losses = 0
min_profit_losses = 0
avg_profit_losses = 0
total_profit_losses = 0
total_change = 0
count = 0
change = 0
months=[]
# Analyzing the csv input data and making decisions
for profit in profitLosses:
    
    total_profit_losses += profit
    count += 1
    if min_profit_losses == 0:
        min_profit_losses = profit
    elif profit > max_profit_losses:
        max_profit_losses = profit
    elif max_profit_losses < min_profit_losses :
        min_profit_losses = profit
    months.append(count)
    
    change_list=[]
    for i in range(86):
        if i == 0:
            change = 0
            change_list.append(change)
        elif i > 0:                
            change = profitLosses[i]-profitLosses[i-1]
            i += 1
            change_list.append(change)
                       
        total_change += change
# Locating the greatest increase/decrease index
x=change_list.index(max(change_list))
y=change_list.index(min(change_list))
#print(x, y)
# Displaying the greatest increase/decrease       
print('The greatest increase was on '+dates[x]+' with an amount of '+str(max(change_list)))
print('The greatest decrease was on '+dates[y]+' with an amount of '+str(min(change_list)))

#print('The list of changes in the consecutive months are: ')
#print(change_list)
# Calculations: Average change, number of monthes, total profit        
total_months=max(months)  
total = total_profit_losses
avg_profit_losses = round(total_profit_losses /count,2)
avg_total_change = round(sum(change_list)/(i-1),2)
# output print statements
print('The total profit loss is '+str(total_profit_losses))
print('The total number of months in the given data is '+str(max(months)))
print('The maximum profit was '+str(max_profit_losses))
print('The average of the profit losses is '+str(avg_profit_losses))
print('The average total change in profit losses is '+str(avg_total_change))
# Exporting the final analysis output to text file
output_file=open("Output.txt", "w")
text_list=['Financial Analysis: ', '------------------------------','Total Months: '+str(total_months), 'Total: '+str(total), 'Average change: '+str(avg_total_change), 'Greatest Increase in Profits: '+dates[x]+' '+str(max(change_list)), 'Greatest decrease in Profits: '+dates[y]+' '+str(min(change_list))]
# Open a text file for the output and write what is in the text list
output_file = open("output.txt", "w")
for line in text_list:
  # write line to output file
    output_file.write(line)
    output_file.write("\n")
output_file.close()
```
